```python
# For Visual Studio:
# %matplotlib inline

# For Jupyer Notebook
%matplotlib notebook

# Import the following dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics

# Load in CSV
pyber_ride_df = pd.read_csv("Resources/Pyber_ride_data.csv")
pyber_ride_df
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data:
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
## 5.1.10 SKILL DRILL - Using the data from the PyBer ride DataFrame, create a bar chart with the following annotations and characteristics:
# - Add error bars.
# - Add caps to the error bars.
# - Change the color of the bars to sky blue.
# - Rotate the labels on the x-axis to horizontal.
# X- Set the y-axis increment to every $5.

# Get the Standard Devision from the "Avg Fare Values Column", which is used for the Error Bars below:
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])

# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data:
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color='skyblue', kind='bar')
plt.xticks(tick_locations, pyber_ride_df["Month"], rotation='horizontal')
plt.yticks(np.arange(0, 51, step=5.0))

# Adds Error Bars and Caps:
plt.errorbar(pyber_ride_df["Month"],
             pyber_ride_df["Avg. Fare ($USD)"],
             yerr=stdev,
             capsize=2,
             linestyle='')

plt.show()
```
